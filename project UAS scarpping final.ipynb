```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_autoinstaller
```

```python
chromedriver_autoinstaller.install()
browser = webdriver.Chrome()
```

```python
browser.get('https://www.lamudi.co.id/east-java/surabaya/house/buy/')
```

```python
soup=BeautifulSoup(browser.page_source,'lxml')
```

```python
rows=soup.find_all('div',class_='ListingCell-AllInfo ListingUnit')
```

```python
data = []
for index in range (1,100) :

    for row in rows:
        nama_element = row.find(class_='ListingCell-KeyInfo-title')
        nama = nama_element.text.replace('\n', ' ').strip() if nama_element is not None and nama_element.text.strip() != "" else None

        location_element = row.find(class_='ListingCell-KeyInfo-address-text')
        location = location_element.text.replace('\n', ' ').strip() if location_element is not None and location_element.text.strip() != "" else None
     
        description_element = row.find(class_='ListingCell-shortDescription')
        description = description_element.text.replace('\n', ' ').strip() if description_element is not None and description_element.text.strip() != "" else None

        price_element = row.find(class_='PriceSection-FirstPrice')
        price = price_element.text.replace('\n', ' ').strip() if price_element is not None and price_element.text.strip() != "" else None
    
        jumlah_kamar_element = row.find(class_='KeyInformation-value_v2 KeyInformation-amenities-icon_v2')
        jumlah_kamar = jumlah_kamar_element.text.replace('\n', ' ').strip() if jumlah_kamar_element is not None and jumlah_kamar_element.text.strip() != "" else None
        
        dt = {
            'nama': nama,
            'location': location,
            'description': description,
            'price': price,
            'jumlah_kamar': jumlah_kamar,
        }

        data.append(dt)
```

```python
df=pd.DataFrame(data)
```
